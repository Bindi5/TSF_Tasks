<span style="color:#730099; font-size:17px"><b>Importing Necessary Libraries</b></span>

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from IPython.display import Image
import io
```

<span style="color:#730099; font-size:17px"><b>Load the 'iris' Dataset and Split it into input & target variable</b></span>

<span style="color:black; font-size:15px"><b>Then create a dataframe with the Input data</b></span>

```python
iris = load_iris()

inp = iris.data
target = iris.target

df = pd.DataFrame(inp, columns = iris.feature_names)
df
```

<span style="color:#730099; font-size:17px"><b>The Dataset Column names and the possible Target Class names</b></span>

```python
print(iris.feature_names)
print(iris.target_names)
```

<span style="color:#730099; font-size:17px"><b>Statistical description about the Dataset</b></span>

```python
df.describe()
```

<span style="color:#730099; font-size:17px"><b>Now again Split the data into Training and Test sets</b></span>

<span style="color:black; font-size:15px"><b>75% of the data into Training set and 25% into Test set</b></span>

```python
x_train, x_test, y_train, y_test = train_test_split(df[iris.feature_names], target, test_size = 0.25, random_state = 0)
```

<span style="color:#730099; font-size:17px"><b>Scikit-learn 4-Step Modeling Pattern</b></span>

<span style="color:black; font-size:17px"><b>Step 1: </b></span><span style="color:#008080; font-size:17px"><b> Import the model you want to use (DecisionTreeClassifier , which is imported earlier)</b></span>

<span style="color:black; font-size:17px"><b>Step 2: </b></span><span style="color:#008080; font-size:17px"><b> Make an instance of DecisionTreeClassifier()</b></span>

<span style="color:black; font-size:17px"><b>Step 3: </b></span><span style="color:#008080; font-size:17px"><b> Fit it to Train the model</b></span>

<span style="color:black; font-size:17px"><b>Step 4: </b></span><span style="color:#008080; font-size:17px"><b> Predict with the test data</b></span>

```python
clf = DecisionTreeClassifier()
 
clf.fit(x_train, y_train)

clf.predict(x_test)
```

<span style="color:#730099; font-size:17px"><b>Model Accuracy</b></span>

```python
clf.score(x_test, y_test)
```

<span style="color:#730099; font-size:17px"><b>Visualize Decision Trees using Matplotlib</b></span>

<span style="color:#008080; font-size:17px"><b>As of scikit-learn version 21.0, Decision Trees can now be plotted with matplotlib using scikit-learnâ€™s <span style="color:#730099"><b>tree.plot_tree </b></span><span style="color:#008080; font-size:17px"><b> without relying on the dot library.</b></span>

```python
plt.figure(figsize=(15,10))
tree.plot_tree(clf,
               feature_names = iris.feature_names,
               class_names = iris.target_names,
               filled = True)
```

<span style="color:#730099; font-size:17px"><b>Print its text representation</b></span>

```python
text = tree.export_text(clf)
print(text)
```

<span style="color:#730099; font-size:17px"><b>Visualize Decision Tree using Graphviz</b></span>

<span style="color:#008060; font-size:16px"><b>Graph visualization is a way of representing structural information as diagrams. The first part of this process involves creating a dot file which is a Graphviz representation of a decision tree.</b></span>

<span style="color:#008090; font-size:17px"><b>* So to plot the tree, first we need to export it to DOT format with </b></span><span style="color:#730099; font-size:16px"><b>export_graphviz </b></span><span style="color:#008090; font-size:17px"><b>method.</b></span>

<span style="color:#008090; font-size:17px"><b>* Then Visualize using pyplotplus </b></span>

```python
dot_data = io.StringIO()

tree.export_graphviz(clf,
                    out_file = dot_data,
                    feature_names = iris.feature_names,
                    class_names = iris.target_names,
                    filled = True)



import pydotplus


graph = pydotplus.graphviz.graph_from_dot_data(dot_data.getvalue())

Image(graph.create_png())
```
